# RegEx (Regular Expression)
- 구조화 되지 않은 일반 텍스트 내에서 **특정 문자열 패턴**의 검색  
  (Ex. 신문기사 / 블로그 텍스트 내 문자열 검색)  
  (Ex. 간단한 xml, html 문서 검색)

$~$

- 정해진 기호를 활용한 문자열 패턴 작성
  - '.': 하나의 문자 (모든 종류의 문자 가능)
  - '*': 앞서 작성된 패턴의 0번 또는 다수 반복
  - '+': 앞서 작성된 패턴의 1번 또는 다수 반복
  - '?': 앞서 작성된 패턴의 0번 또는 1번 발생
  - '^': 대상 문자열의 시작
  - '$': 대상 문자열의 끝
  - '[]': 사용가능한 패턴의 집합(set), 반전 ^  
    (ex. [abcd]: a, b, c, d 중 어떤 것이 와도 가능)  
    (ex. [^abcd]: a, b, c, d가 아닌 것만 가능)
  - '{m,n}': 앞서 작성된 패턴의 m번이상, n번 이하 반복

```python
import re
```

- **re.search('대상 문자열', text)**
- if there isn't exist, return None

```python
result = re.search('컨텐츠', naver_html.text)
print(result.group())
```

```python
text = 'abcdef'

def exist(result):
  if result != None:
    print(result.group())
  else:
    print('Not found')
  
re1 = re.search('a', text)
re2 = re.search('.e', text)
re3 = re.search('a*', text)
re4 = re.search('a.*', text)
re5 = re.search('f.*', text)
re6 = re.search('f.+', text)
re7 = re.search('f.?', text)
re8 = re.search('a.?', text)
re9 = re.search('[bcde]', text)
re10 = re.search('[^bcde]', text)

exist(re1)
exist(re2)
exist(re3)
exist(re4)
exist(re5)
exist(re6)
exist(re7)
exist(re8)
exist(re9)
exist(re10)
```

```python
text = 'abcdef'

re1 = re.search('^.b', text)
re2 = re.search('^.*c', text)
re3 = re.search('^.?c', text)
re4 = re.search('d.$', text)
re5 = re.search('d..$', text)
re6 = re.search('^.*c.+$', text)

exist(re1)
exist(re2)
exist(re3)
exist(re4)
exist(re5)
exist(re6)
```

```python
text = 'abcdef'

re1 = re.search('^[abcdef]$', text) # 하나의 문자 (a,b,c,d,e,f) 중
re2 = re.search('^[abcdef]*$', text)
re3 = re.search('^[abcdef]+$', text)

exist(re1)
exist(re2)
exist(re3)
```

```python
text = 'aaaabbbbbbcccdeef'

re1 = re.search('[abcdef]{3}', text)
re2 = re.search('a{3}', text)
re3 = re.search('a{5}', text)
re4 = re.search('a{4}', text)
re5 = re.search('a{2, 4}', text)

exist(re1)
exist(re2)
exist(re3)
exist(re4)
exist(re5)
```

```python
result = re.findall('a.{3}', naver_html.text)
print(result)

result = re.findall('.{3}[a-z]<', naver_html.text)
print(result)

result = re.findall('컨.{2}', naver_html.text)
print(result)

result = re.findall('코.{4}', naver_html.text)
print(result)
```
